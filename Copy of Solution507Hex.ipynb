```python
import random
import matplotlib.pyplot as plt


OrigCoords = [(2,0), (1,1.732),  (-1,1.732), (-2,0), (-1, -1.732), (1, -1.732)] # These are the corners of the hexagon
OrigX = [x for x,y in OrigCoords]
OrigY = [y for x,y in OrigCoords]
OrigX.append(OrigX[0])
OrigY.append(OrigY[0])

def drawpoints(Points):
  Xlist = [x for x,y in Points]
  Ylist = [y for x,y in Points]
  fig, ax = plt.subplots(figsize=(8, 8))
  ax.set_facecolor('azure')
  ax.scatter(Xlist, Ylist,2)
  ax.plot(OrigX,OrigY, 'r')
  plt.show()

```
